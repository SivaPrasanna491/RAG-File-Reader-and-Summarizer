```python
## web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader(file_path='Unit 5 pdf.pdf')
```

```python
text_documents = loader.load()
```

```python
text_documents
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(text_documents)
```

```python
docs
```

```python
## vector embeddings and vector store
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
db = Chroma.from_documents(documents=docs[:5], embedding=OllamaEmbeddings(model='nomic-embed-text:v1.5'))
```

```python
query = "Explain me about this pdf. What is it saying"
result = db.similarity_search(query=query, k = 10)
```

```python
result
```

```python
result[0].page_content
```

```python
from langchain_community.llms import Ollama
```

```python
llm = Ollama(model='mistral')
```

```python
llm
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template("""
Answer the following question based only on the provided context. 
Think step by step before providing a detailed answer. 
I will tip you $1000 if the user finds the answer helpful. 
<context>
{context}
</context>
Question: {input}""")
```

### Chain Introduction

```python
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
```

```python
document_chain = create_stuff_documents_chain(llm=llm, prompt=prompt)
```

```python
document_chain
```

```python
retriever = db.as_retriever()
```

```python
retriever
```

### creating Retriever Chain

```python
from langchain_classic.chains import create_retrieval_chain
```

```python
retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
response = retrieval_chain.invoke({
    "input": "Explain me this document. What is it saying",
    })
response['answer']
```
