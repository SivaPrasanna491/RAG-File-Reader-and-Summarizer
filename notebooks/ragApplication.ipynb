```python
from langchain_community.document_loaders import TextLoader, WebBaseLoader, PyPDFLoader, UnstructuredExcelLoader
```

```python
import bs4
```

```python
text_loader = TextLoader(file_path='speech.txt', encoding='utf-8')
pdf_loader = PyPDFLoader(file_path='Unit 5 pdf.pdf')
excel_loader = UnstructuredExcelLoader(file_path='RTPof VAL 2nd HYS-2024_17-04-25.xlsx', mode="elements")
```

```python
text_documents = text_loader.load()
pdf_documents = pdf_loader.load()
excel_documents = excel_loader.load()
```

### dividing the documents into chunks

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 5000, chunk_overlap = 500)
```

```python
text_docs = splitter.split_documents(text_documents)
pdf_docs = splitter.split_documents(pdf_documents)
excel_docs = splitter.split_documents(excel_documents)
```

```python
text_docs
```

```python
excel_docs
```

```python
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import Chroma, FAISS, LanceDB
```

```python
chroma_db = Chroma.from_documents(documents=text_docs[:5], embedding=OllamaEmbeddings(model='nomic-embed-text:v1.5'))
faiss_db = FAISS.from_documents(documents=pdf_docs[:5], embedding=OllamaEmbeddings(model='snowflake-arctic-embed:335m'))
lance_db = LanceDB.from_documents(documents=excel_docs[:5], embedding=OllamaEmbeddings(model='nomic-embed-text:v1.5'))
```

```python
query = "Explain me the concept of python according to this pdf document"
result = faiss_db.similarity_search(query=query, k = 10)
result
```

```python
query = "One Piece is it real?"
result = chroma_db.similarity_search(query=query)
```

```python
print(f"{result[1].page_content}\n")
```

### Integrating Chain and Retriever

```python
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template("""
You are a helpful assistant. Answer the user query according to the context.
Think step by step before answering the user query and don't generate random
answers like a stupid think twice and then answer the user query according to
the user query.I will give you credits if you answer the user query correctly.
<context>
{context}
</context>
Question: {input}""")
```

```python
from langchain_community.llms import Ollama
```

```python
llm = Ollama(model='qwen3:0.6b')
```

```python
document_chain = create_stuff_documents_chain(llm = llm, prompt = prompt)
```

```python
document_chain
```

## creating the retriever for particular vector store

```python
chroma_retriever = chroma_db.as_retriever()
faiss_retriever = faiss_db.as_retriever()
lance_retriever = lance_db.as_retriever()
```

```python
chroma_retriever
```

## Combining both retriever and chain

```python
from langchain_classic.chains import create_retrieval_chain
```

```python
chroma_chain = create_retrieval_chain(chroma_retriever, document_chain)
faiss_chain = create_retrieval_chain(faiss_retriever, document_chain)
lance_chain = create_retrieval_chain(lance_retriever, document_chain)
```

```python
response = lance_chain.invoke(
    {
        "input": "Explain me what the excel document is saying or trying to explain."
    }
)
print(f"{response['answer']} \n")
```

```python
response = chroma_chain.invoke(
    {
        "input": "What is One Piece and who is Monkey D Luffy"
    }
)
response['answer']
```
